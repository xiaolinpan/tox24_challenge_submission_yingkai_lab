```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# 假设您有两个数据集的 SMILES 列表（这里以例子代替）

def smiles_to_ecfp(smiles_list, radius=1, nBits=4096):
    """将SMILES列表转换为ECFP指纹矩阵。"""
    fingerprints = []
    for smi in smiles_list:
        mol = Chem.MolFromSmiles(smi)
        if mol is not None:
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
            arr = np.zeros((1,), dtype=np.int8)
            Chem.DataStructs.ConvertToNumpyArray(fp, arr)
            fingerprints.append(arr)
        else:
            print(f"Warning: could not parse SMILES {smi}")
    return np.array(fingerprints)
```

```python
combined_fp.shape
```

```python
# 将两个数据集的指纹分别算出
fp_dataset1 = smiles_to_ecfp(dataset1, radius=1, nBits=4096)
```

```python


fp_dataset2 = smiles_to_ecfp(dataset2, radius=1, nBits=4096)

# 合并两个数据集的指纹用于统一降维
combined_fp = np.concatenate([fp_dataset1, fp_dataset2], axis=0)
```

```python
# 使用 t-SNE 进行降维
# tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=10000)
tsne = PCA(n_components=2, random_state=42)
embedding = tsne.fit_transform(combined_fp)

```

```python
# 根据数据集标签区分，前半部分是dataset1, 后半部分是dataset2
n1 = len(fp_dataset1)
n2 = len(fp_dataset2)

x = embedding[:, 0]
y = embedding[:, 1]

# 绘图
plt.figure(figsize=(8,6))
plt.scatter(x[:n1], y[:n1], c='blue', label='Dataset 1')
plt.scatter(x[n1:], y[n1:], c='red', label='Dataset 2')
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.title("t-SNE on Two Datasets using ECFP 2048")
plt.legend()
```
